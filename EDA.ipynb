# preprocessing

# check for duplicate values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# Dealing with null values

```python
df.isnull().sum()
```

# if number of missing values is large wrt total values then dropping the column is the suitable strategy

```python
df.drop('Cabin',axis=1,inplace=True)
```

# if the numeric column present most of  the null value we replace it by mean or meadian 

```python
avg_age=df.Age.mean()
```

```python
df.Age.replace(np.nan,avg_age,inplace=True) #replace the null value by average(mean)
```

```python
df.isnull().sum()
```

```python
freq_port=df.Embarked.dropna().mode()[0] 
df.replace(np.nan,freq_port,inplace=True)
```

```python
df.isnull().sum()
```

```python
# To compare dead and survived
plt.figure(figsize=(9,5),dpi=100)
sns.countplot(x='Survived',data=df)
```

```python
df.Survived.value_counts()
```

```python
Survived_data=df.Survived.value_counts(normalize=True)
```

```python
print('Percentage of Survived',Survived_data[1],'%')
print('Percentage of Unurvived',Survived_data[0],'%')
```

```python
#to compare dead and survived wrt sex\
plt.figure(figsize=(9,5),dpi=100)
sns.countplot(x='Survived',data=df,hue='Sex')
```

```python
male_female=df.Sex.value_counts()
male_female
```

```python
male_survived=len(df[(df.Sex=='male')& (df.Survived==1)])
male_survived
```

```python
female_survived=len(df[(df.Sex=='female')& (df.Survived==1)])
female_survived
```

```python
#to compare dead and survived wrt age\
plt.figure(figsize=(9,5),dpi=100)
sns.countplot(x='Survived',data=df,hue='Age')
```

```python
male_surviver_rate=(male_survived/male_female['male'])*100
female_surviver_rate=(female_survived/male_female['female'])*100
print("Rate of male survived=",male_surviver_rate)
print("Rate of female survived=",female_surviver_rate)
```

```python
svl_rate=df.groupby('Sex')['Survived'].value_counts(normalize=True)
svl_rate
```

```python
print("percentage of female survivors:",svl_rate['female'][1]*100)
print("percentage of female unsurvivors:",svl_rate['female'][0]*100)
print("percentage of male survivors:",svl_rate['male'][1]*100)
print("percentage of male unsurvivors:",svl_rate['male'][0]*100)
```

```python
#Travellers is passenger classes
plt.figure(figsize=(9,5),dpi=100)
sns.countplot(x='Pclass',data=df)
```

```python
plt.figure(figsize=(9,5),dpi=100)
sns.countplot(x='Pclass',data=df,hue='Survived')
```

```python
svl_rate=df.groupby('Pclass')['Survived'].value_counts(normalize=True)
svl_rate
```

```python
print('Percentage of 1st Passenger Class Survivors     : ', svl_rate[1][1]*100)
print('Percentage of 1st Passenger Class Non-Survivors : ', svl_rate[1][0]*100)
print('Percentage of 2nd Passenger Class Survivors     : ', svl_rate[2][1]*100)
print('Percentage of 2nd Passenger Class Non-Survivors : ', svl_rate[2][0]*100)
print('Percentage of 3rd Passenger Class Survivors     : ', svl_rate[3][1]*100)
print('Percentage of 3rd Passenger Class Non-Survivors : ', svl_rate[3][0]*100)
```

```python
pclass_svl=[svl_rate[1][1],svl_rate[2][1],svl_rate[3][1]]
plt.figure(figsize=(9,5),dpi=100)
plt.pie(pclass_svl,labels = ['Class1','Class2','Class3'], autopct = '%1.1f%%')
```

```python
fig = plt.figure(figsize = (9,4), dpi = 200)
p1 = fig.add_subplot(1,2,1)
p2 = fig.add_subplot(1,2,2)

pclass_svl = [svl_rate[1][1],svl_rate[2][1],svl_rate[3][1]]
pclass_non_svl = [svl_rate[1][0],svl_rate[2][0],svl_rate[3][0]]

classes = ['Class1','Class2','Class3']

p1.pie(pclass_svl, labels = classes, autopct = '%1.1f%%') #autopct -to show percentages on the slices
p2.pie(pclass_non_svl, labels = classes, autopct = '%1.1f%%')
p1.title.set_text('Survival proportions')
p2.title.set_text('Death Proportions')
```
